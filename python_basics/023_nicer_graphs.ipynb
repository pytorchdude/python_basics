```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# the following is done within a single axis area
x = np.linspace(-3,3,101)

plt.plot(x,x,label='y=x')
plt.plot(x,x**2,label='y=x**2')
plt.plot(x,x**3,label='y=x**3')

plt.legend()
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.title('A sweet plot of f of x')
plt.xlim(x[0],x[-1])                  # by defualt takes a tuple suchas (-3,3)
plt.ylim(-10,10) #this way adjusts as the graph changes -- keepinig a ratio of sorts

plt.gca().set_aspect(.3) # corresponds to value to ratio of the distance vs each axis between axis values
                             # can do 'auto' or a number such as 1 or .1 or .3 or 1.5 -- need to play with it for 
                             # individual needs                        

plt.show()
```

```python
# the differences for plot vs subplot to make look good
fig,ax = plt.subplots(1)

# we need to replace all plt. with ax.
ax.plot(x,x,label='y=x')
ax.plot(x,x**2,label='y=x**2')
ax.plot(x,x**3,label='y=x**3')

# uses set_then attriobute
ax.legend()
ax.set_xlabel('x')                
ax.set_ylabel('y = f(x)')
ax.set_title('A sweet plot of f of x')
ax.set_xlim(x[0],x[-1])                  # by defualt takes a tuple suchas (-3,3)
ax.set_ylim(-10,10) #this way adjusts as the graph changes -- keepinig a ratio of sorts

ax.set_aspect('auto') # there is no gca attribute for ax. -- corresponds to value to ratio of the distance vs each axis between axis values
                             # can do 'auto' or a number such as 1 or .1 or .3 or 1.5 -- need to play with it for 
                             # individual needs                        

plt.show()
```

```python
# the differences for plot vs subplot to make look good
fig,ax = plt.subplots(2,3, figsize=(15,7)) # can do for multiple subplots

# we need to replace all plt. with ax.
ax[0,2].plot(x,x,label='y=x')
ax[1,2].plot(x,x**2,label='y=x**2')
ax[0,1].plot(x,x**3,label='y=x**3')

# uses set_then attriobute
ax[0,2].legend()
ax[1,2].set_xlabel('x')                
ax[0,1].set_ylabel('y = f(x)')
ax[0,1].set_title('A sweet plot of f of x')
ax[0,2].set_xlim(x[0],x[-1])                  # by defualt takes a tuple suchas (-3,3)
ax[0,2].set_ylim(-10,10) #this way adjusts as the graph changes -- keepinig a ratio of sorts

#ax.set_aspect('auto') # there is no gca attribute for ax. -- corresponds to value to ratio of the distance vs each axis between axis values
                             # can do 'auto' or a number such as 1 or .1 or .3 or 1.5 -- need to play with it for 
                             # individual needs                        
plt.tight_layout()  # perfect for this
plt.show()
```

```python
# lets create code to create something cool
fig,ax = plt.subplots()

x = np.linspace(0,10,100)

for i in np.linspace(0,1,50):
    ax.plot(x,x**i,color=[i/2,0,i]) # this color method uses 3 values, RGB - red, green, and blue

plt.show()
```
