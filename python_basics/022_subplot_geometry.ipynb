```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.subplot(1,2,1)             # 1 row and 2 columns and 3rd input is the index to which we are plotting
plt.plot(np.random.randn(10))

plt.subplot(1,2,2)
plt.plot(np.random.randn(10))

plt.show

# 1st subplot is on the left side and the 2nd subplot is on the right side
```

```python
# subplot indexing works normal with 0 as the 1st index

plt.subplot(2,2,1)             # 1 row and 2 columns and plts to 1st index
plt.plot(np.random.randn(10))

plt.subplot(2,2,2)             # 1 row and 2 columns and plts to 2nd index
plt.plot(np.random.randn(10))

plt.subplot(2,2,3)             # 1 row and 2 columns and plts to 3rd index
plt.plot(np.random.randn(10))


plt.subplot(2,2,4)             # 2 rows, 2 columns, and plts to 4th index 
plt.plot(np.random.randn(10))

plt.show

# 1st subplot is on the left side and the 2nd subplot is on the right side top(blank)
# 3rd subplot in blank lower left and 4th is the lower right
```

```python
# generating subplots
fig,ax = plt.subplots(1,2)  # 2 inputs figure and axis 

x = np.arange(10)

ax[0].plot(x,x**2, 'b')
ax[0].plot(x,x, 'k')          # k is black
ax[1].plot(x,np.sqrt(x), 'r')

plt.show
```

```python
# generating subplots
fig,ax = plt.subplots(1,2, figsize=(15,3))  # 2 inputs figure and axis with 3rd optional input as a tuple width, height

x = np.arange(10)

ax[0].plot(x,x**2, 'b')
ax[0].plot(x,x, 'k')          # k is black
ax[1].plot(x,np.sqrt(x), 'r')

plt.show()
# the figsize creates the size of the grpah area indexes
```

```python
# generating subplots
fig,ax = plt.subplots(2,2)  # creats a 2 x 2 subplot area of 4 indexes

x = np.arange(10)

ax[0,0].plot(np.random.randn(4,4))
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4))
ax[1,1].plot(np.random.randn(4,4))

plt.tight_layout()                  # makes sure there is no overlapping
plt.show()
# the figsize creates the size of the grpah area indexes
# 0,0 is top left 0,1 is top right, 1,0 is bottom left and 1,1 is bottom right
```

```python
fig,ax = plt.subplots(3,3,figsize=(7,7))

ax.shape
```

```python
fig,ax = plt.subplots(3,3,figsize=(7,7))

ax.flatten().shape # converts to 9 element matrix
```

```python
fig,ax = plt.subplots(3,3,figsize=(7,7))

for i in ax.flatten():                    # flattens for each iteration
    i.plot(np.random.randn(3,9))          # creates 3x9 matrix

plt.show()    
```
