```python
plt.plot(3,4, 'ro') # red dot -- with letter in fron of symbol request
plt.show()
plt.plot(2,4.2, 'b^') # blue triangle -- with letter in fron of symbol request
# these are plotted in different axis graphs
```

```python
plt.plot(3,4, 'ro', label='red dot') # red dot -- with letter in fron of symbol request
plt.plot(2,4.2, 'b^', label='blue triangle') # blue triangle -- with letter in fron of symbol request
plt.show()
# these are plotted inn the same axis -- we attach labels -- but we need to display them as a legend
```

```python
plt.plot(2,4.2, 'b^', label='blue triangle') # blue triangle -- with letter in fron of symbol request
plt.plot(3,4, 'ro', label='red dot') # red dot -- with letter in fron of symbol request
plt.legend()                         # displays the legend
plt.show()
# these are plotted inn the same axis -- we attach labels -- and use plot.legend() to display the legend
# legend display in order of instruction
```

```python
plt.plot(2,4.2, 'b^', label='blue triangle') # blue triangle -- with letter in fron of symbol request
plt.plot(3,4, 'ro', label='red dot') # red dot -- with letter in fron of symbol request
plt.legend(['blue','red'])                         # displays the legend - with customs -- still in order of instruction
plt.show()
# these are plotted inn the same axis -- we attach labels -- and use plot.legend() to display the legend
# legend display in order of instruction
```

```python
# plotting lines
x = [0,1]
y = [0,2]

plt.plot(x,y,'m') # magenta line
plt.show()
```

```python
x = [0,1,3]
y = [0,2,-1]

plt.plot(x,y,'m') # magenta lines
plt.show()
```

```python
x = [0,1,3]
y = [0,2,-1]

plt.plot(x,y,'ms') # magenta squares -- python/matplotlib knows
plt.show()
```

```python
x = [0,1,3]
y = [0,2,-1]

plt.plot(x,y,'ms-') # magenta squares -- lets connect the squares w/ solid line
plt.show()
```

```python
x = [0,1,3]
y = [0,2,-1]

plt.plot(x,y,'ms--') # magenta squares -- lets connect the squares w/ dashed line
plt.show()
```

```python
x = [0,1,3]
y = [0,2,-1]

plt.plot(x,y,'ms:') # magenta squares -- lets connect the squares w/ dotted line
plt.show()
```
