```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg') 
# import htis to create higher resolution svg vector draws not jpg
```

```python
x = np.linspace(.5,5,10)
y1 = np.log(x)
y2 = 2-np.sqrt(x)

plt.plot(x,y1,'bo-', label='log')
plt.plot(x,y2, 'rs-', label='sqrt')
plt.legend()

plt.show()
plt.savefig('test.png') # will save locally in Jupyter or will save local to the google colab drive that you can connect
# your google drive using:
# from google.colab import files
# files.download('test.png')
```
