```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import imageio.v2 as imageio
```

```python
# visulize a matrix
m = 3
n = 5

M = np.random.randint(10,size=(m,n))
print(M)

plt.imshow(M)
plt.colorbar()
plt.show()
```

```python
# let's make it look better and make it useful
m = 3
n = 5

M = np.random.randint(10,size=(m,n))
print(M)

plt.imshow(M)

for i in range(m):
    for j in range(n):
        plt.text(j,i,str(M[i,j]), fontsize=20, 
                 horizontalalignment='center', 
                 verticalalignment='center')

plt.colorbar()
plt.show()
```

```python
# Visualize the Hilbert matrix using seaborn with heatmap
# H is 10x10

n = 10
H = np.zeros((n,n))

for i in range(n):              
    for j in range(n):
        H[i,j] = 1 / ((i+1)+(j+1)-1) # python starts couinting at 0 so add 1 to each i and j or just i+j+1 

sns.heatmap(H, vmax=.4)
plt.show()
```
